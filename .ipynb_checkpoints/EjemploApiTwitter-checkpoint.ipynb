```python
tweets = []
def username_tweets_to_csv(username,count):
    try:   
        #Crea un m√©todo de consulta usando par√°metros
        tweets = tweepy.Cursor(api.user_timeline,id=username).items(count)
        #Extrae informaci√≥n de tweet iterativamente
        tweets_list = [[tweet.created_at, tweet.id, tweet.text] for tweet in tweets]
        # Crea un dataframe desde las listas de tweets        
        tweets_df = pd.DataFrame(tweets_list,columns=['Datetime', 'Tweet Id', 'Text'])
        # Convertiendo dataframe a CSV 
        tweets_df.to_csv('{}-tweets.csv'.format(username), sep=',', index = False)

    except BaseException as e:
          print('failed on_status,',str(e))
          time.sleep(3)
```

```python
# Input username to scrape tweets and name csv file
# Max recent tweets pulls x amount of most recent tweets from that user
username = 'repsol'
count = 20

# Llama a funcipara pasar count de twitter en un archivo CSV
username_tweets_to_csv(username, count)
```
